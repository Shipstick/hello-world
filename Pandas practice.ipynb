```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df[(df['location'] == 'Russian Federation') & (df['cause'] == 
'HIV/AIDS and sexually transmitted infections') & (df['sex'] == 'Both')]
```

```python
df[(df['location'] == 'Russian Federation') & (df['cause'] == 
'HIV/AIDS and sexually transmitted infections') & (df['sex'] == 'Both')].pivot_table(index = 'year',
  columns = 'sex', values = 'value', aggfunc = 'mean')
```

```python
df[(df['location'] == 'Russian Federation') & (df['cause'] == 
'HIV/AIDS and sexually transmitted infections') & (df['sex'] == 'Both')].pivot_table(index = 'year',
                              columns = 'sex', values = 'value', aggfunc = 'mean').plot(kind = 'line')
plt.ylabel('deaths per 100 000 people')
plt.title('deaths in Russia from HIV/AIDS and sexually transmitted infections')
```

```python
df_2019 = df[(df['sex'].isin(['Male','Female'])) & (df['year'] == 2019)].pivot_table(values = 'value', 
     aggfunc = 'mean', index = 'cause', columns = 'sex')
df_2019.plot(kind = 'bar')
plt.ylabel('deaths per 100 000 people')
plt.title('deaths by sex')
```
