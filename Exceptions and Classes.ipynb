```python
class Toyota:
    def __init__(self): #
        self.color = 'Индиго'
        self.price = '1 000 000 руб'
        self.max_velocity = '200 км/ч'
        self.current_velocity = '0 км/ч'
        self.engine_rpm = 0

    def start(self):
        print('Мотор запущен')
        self.engine_rpm = 900

    def go(self):
        print('Поехали!')
        self.engine_rpm = 2000
        self.current_velocity = '20 км/ч'
my_car = Toyota()
print('color',my_car.color)
print('price',my_car.price)
print('max_velocity',my_car.max_velocity)
print('rpm',my_car.engine_rpm)
print('current_velocity',my_car.current_velocity)

my_car.start()
print('rpm',my_car.engine_rpm)
my_car.go()
print('rpm',my_car.engine_rpm)
print('current_velocity',my_car.current_velocity)
```

```python
class Backpack:
    def __init__(self):
        self.content =[]
    def add(self,item):
        self.content.append(item)
        print(u'В рюкзак положили: ',item)
    def inspect(self):
        print(u'в рюкзаке лежит:')
        for item in self.content:
            print('    ', item)
my_backpack = Backpack()
my_backpack.add(item='ноутбук')
my_backpack.add(item='зарядка для ноутбука')
my_backpack.inspect()
```

```python
from random import randint


class Man:

    def __init__(self, name):
        self.name = name
        self.fullness = 50
        self.house = None

    def __str__(self):
        return 'Я - {}, сытость {}'.format(
            self.name, self.fullness)

    def eat(self):
        if self.house.food >= 10:
            print('{} поел'.format(self.name))
            self.fullness += 10
            self.house.food -= 10
        else:
            print('{} нет еды'.format(self.name))

    def shopping(self):
        if self.house.money >= 50:
            print('{} cходил в магазин за едой'.format(self.name))
            self.house.money -= 50
            self.house.food += 50
        else:
            print('{} деньги кончились! '.format(self.name))
            self.work()

    def kit_shop(self):
        if self.house.money >= 10 and self.house.kit_kat <= 10:
            print('{} cходил в магазин за едой кошака'.format(self.name))
            self.house.money -= 10
            self.house.kit_kat += 30
        else:
            print('{} Коту нечего есть и денег нет! '.format(self.name))
            self.work()


    def go_in_the_house(self, house):
        self.house = house
        self.fullness -= 10
        print('{} Заехал в дом'.format(self.name))

    def work(self):
        print('{} cходил на работу'.format(self.name))
        self.house.money += 50
        self.fullness -= 10

    def watch_MTV(self):
        print('{} смотрел MTV целый день'.format(self.name))
        self.fullness -= 10

    def Kenny_kill(self):
        print('{} убил Кенни! '.format(self.name))

    def act(self):
        if self.fullness <= 0:
            print('{} умер... '.format(self.name))
            return
        dice = randint(1, 8)
        if self.fullness < 20:
            self.eat()
        elif self.house.kit_kat <= 10:
            self.kit_shop()
        elif self.house.food < 10:
            self.shopping()
        elif self.house.money < 50:
            self.work()
        elif dice == 1:
            self.work()
        elif dice == 2:
            self.eat()
        elif dice == 3 or dice == 4:
            self.Kenny_kill()
        else:
            self.watch_MTV()

class Kit:

    def __init__(self,name):
        self.house = None
        self.golod = 50 #аппетит
        self.name = name
        self.drowsiness = 50 #сонливость (>80 - спать)

    def go_in_the_house(self, house):
        self.house = house
        self.golod -= 10
        print('{} Заехал в дом,аппетит каджита {}'.format(self.name, self.golod))

    def __str__(self):
        return 'Я - {}, сытости {} , сонливости {}'.format(
            self.name, self.golod,self.drowsiness)

    def sleep(self):
        self.drowsiness -= 15
        print('{} Поспал - отдхонул, сонливости {} (-15 к сонливости)'.format(self.name,self.drowsiness))

    def wallpaper_scratch(self):
        self.golod -= 10
        self.drowsiness += 15
        print('{} Царапал обои (-10 к сытости, +15 к сонливости)'.format(self.name))

    def eat_kat(self):
        self.golod += 10
        self.drowsiness += 12
        try:
            self.house.kit_kat -= 10
        except AttributeError as exc:
            print(f'ошибка атрибута {exc}')
        print(f'{self.name} Поел - потолстел (+10 к сытости,+12 к сонливости), \
        Вискаса {self.house.kit_kat}, \
        сытости {self.golod}, \
        сонливость {self.drowsiness}')

    def act(self):
        dice = randint(1, 6)
        if self.golod <= 0:
            print('{} умер... '.format(self.name))
            return
        elif self.golod <= 20:
            self.eat_kat()
        elif self.drowsiness >= 80:
            self.sleep()
        elif dice == 1 or dice == 2:
            self.eat_kat()
        elif  dice == 3 or dice == 4:
            self.sleep()
        elif dice == 5 or dice == 6:
            self.wallpaper_scratch()

class House:

    def __init__(self):
        self.food = 50
        self.money = 0
        self.kit_kat = 50

    def __str__(self):
        return 'В доме еды осталось {}, вискаса {}, денег осталось {}'.format(
            self.food, self.kit_kat ,self.money)

citizens = [
    Man(name='Бивис'),
    Man(name='Батхед'),
    Man(name='Кенни'),
    Kit(name='Раф')
]
my_sweet_home = House()
for citisen in citizens:
    citisen.go_in_the_house(house=my_sweet_home)
for day in range(1, 365):
    print('=========================== день {} ======================='.format(day))
    for citisen in citizens:
        citisen.act()
    print('---------------------------В конце дня-----------------------------')
    for citisen in citizens:
        print(citisen)
    print(my_sweet_home)
```
